# Importing major libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing data

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

# Merging both files into 1

```python
final=pd.merge(movies,credits,left_on='id',right_on='movie_id')
```

```python
final.head()
```

```python
final['original_language'].value_counts()/len(final['original_language'])
```

```python
#Keeping only relevant columns

final=final[['genres','id','keywords','title_x','overview','cast','crew']]
```

```python
#checking if any null values are there

final.isnull().sum()
```

```python
#checking if any duplicate values are there

final.duplicated().sum()
```

```python
#Drop null values

final.dropna(inplace=True)
final.reset_index(inplace=True,drop=True)
```

```python
final.head(1)
```

```python
final.columns=['genre','id','keywords','title','overview','cast','crew']
```

```python
final.head(1)
```

```python
final['genre'][0]
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
#Converting StringList to List

final['genre']=final['genre'].apply(ast.literal_eval)
final['keywords']=final['keywords'].apply(ast.literal_eval)
final['cast']=final['cast'].apply(ast.literal_eval)
final['crew']=final['crew'].apply(ast.literal_eval)
```

```python
final.head()
```

```python
a[0]['name']
```

```python
#Function to fetch out the keywords

def funcname(obj):
    L=[]
    for i in obj:
        L.append(i['name'])
    return L
```

```python
final['genre']=final['genre'].apply(funcname)
final['keywords']=final['keywords'].apply(funcname)

#only top 3 actor/actress are taken

final['cast']=final['cast'].apply(funcname).apply(lambda x: x[0:3])
```

```python
final['cast']
```

```python
#removing the space between firstname and lastname, which will help during countvectorisation

final['cast'] = final['cast'].apply(lambda x: [i.replace(' ','') for i in x])
```

```python
final.head()
```

```python
final['crew'][0]
```

```python
#Function to fetch out the director

def funcjob(obj):
    L=[]
    for i in obj:
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
final['crew']=final['crew'].apply(funcjob)
final['crew']
```

```python
#removing the space between firstname and lastname, which will help during countvectorisation

final['crew'] = final['crew'].apply(lambda x: [i.replace(' ','') for i in x])
```

```python
final.head()
```

```python
final[['genre','keywords']]
```

```python
final['genre'] = final['genre'].apply(lambda x: [i.replace(' ','') for i in x])
final['keywords'] = final['keywords'].apply(lambda x: [i.replace(' ','') for i in x])
```

```python
final=final[['id','title','genre','keywords','overview','cast','crew']]
```

```python
final.sample()
```

```python
#Splitting the overview to a list

final['overview']=final['overview'].apply(lambda x: x.split(' '))
```

```python
final['tags']=final['overview']+final['genre']+final['keywords']+final['cast']+final['crew']
```

```python
final.head()
```

```python
final=final[['id','title','tags']]
```

```python
final
```

# Stemming the tags

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
```

```python
#Stemming the words to decrease the confusion between accident/accidentaly/accidental etc.

final['tags'] = final['tags'].apply(lambda x: [stemmer.stem(i) for i in x])
```

```python
#Converting the list to a string

final['tags']=final['tags'].apply(lambda x: ' '.join(x))
```

```python
final['tags'] = final['tags'].apply(lambda x: x.lower())
```

```python
final.head()
```

```python
final.tags[0]
```

# Count Vectorisation

```python
from sklearn.feature_extraction.text import CountVectorizer
cvector = CountVectorizer(stop_words='english',max_features=5000)
```

```python
#Converting the tags to vectors

vectors=cvector.fit_transform(final['tags']).toarray()
```

```python
vectors
```

```python
cvector.get_feature_names_out()
for w in cvector.get_feature_names_out():
    print (w)
```

# Checking Similarity between different tags

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#checking the cosine similarity between different vectors

similarity = cosine_similarity(vectors)
```

```python
similarity_df=pd.DataFrame(similarity,columns=final['title'],index=final['title'])
similarity_df.head()
```

```python
similarity_df.iloc[1].sort_values(ascending=False)
```
